```python
df1 = pd.read_csv("house_prices.csv")
df1.head(2)
```

```python
df2 = df1.drop(['Title','Sub_title','Post_URL','Seller_name','published_date','Geo_Address','Address','Description'],axis='columns')
df2.head(11)
```

```python
df3=df2.copy()
```

```python

df3["Price"] = df3["Price"].replace("[Rs,]", "", regex=True).astype(int)
df3["House size"] = df3["House size"].replace("[sqft,]", "", regex=True).astype(float)
df3["Land size"] = df3["Land size"].replace("[perches,]", "", regex=True)
df3.head(10)
```

```python
df3.location = df3.Location.apply(lambda x: x.strip())
location_stats = df3['Location'].value_counts(ascending=False)
location_stats
```

```python
df3['Location'] = df3['Location'].apply(lambda y: y.split(',')[0] )
df3.head(2)
```

```python
df4=df3.copy()
#df4['Baths'] = df4['Baths'].astype(int)
df4['Baths'] = pd.to_numeric(df4['Baths'],errors='coerce')
df4['Land size'] = pd.to_numeric(df4['Land size'],errors='coerce')
df4['Beds'] = pd.to_numeric(df4['Beds'],errors='coerce')
df4.dtypes
```

```python
data= df4[['Price','Baths','Land size','Beds','House size']]
```
